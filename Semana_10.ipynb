Faça uma pesquisa sobre como é possível aplicar os operadores de gradiente Roberts, Prewitt e Sobel para detectar bordas em uma imagem, usando a linguagem Python.

Para cada operador, inclua o trecho de código que exemplifica a operação e um conjunto de quatro imagens (para cada operação) mostrando a imagem original, a imagem resultante da operação na horizontal, a imagem resultante da operação na vertical e a imagem representando a magnitude do gradiente.



```python
import numpy as np
import matplotlib.pyplot as plt

from skimage.io import imread
from skimage.filters import roberts, sobel, sobel_h, sobel_v, prewitt, roberts_pos_diag, roberts_neg_diag, prewitt_h, prewitt_v

img = imread("kodim23.png", as_gray=True)
op_roberts = roberts(img)
op_prewitt = prewitt(img)
op_sobel = sobel(img)
```

# Roberts


```python

horizontal = roberts_pos_diag(img)
vertical = roberts_neg_diag(img)

fig, axes = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True, figsize=(10, 10))

ax = axes.ravel()

ax[0].imshow(img, cmap=plt.cm.gray)
ax[0].set_title('Imagem Original')

ax[1].imshow(horizontal, cmap=plt.cm.gray)
ax[1].set_title('Imagem na Horizontal')

ax[2].imshow(vertical, cmap=plt.cm.gray)
ax[2].set_title('Imagem na Vertical')

ax[3].imshow(op_roberts, cmap=plt.cm.gray)
ax[3].set_title('Operador de Roberts')

for a in ax:
    a.axis('off')

plt.tight_layout()
plt.show()
```

# Sobel

```python
horizontal = sobel_h(img)
vertical = sobel_v(img)

fig, axes = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True, figsize=(10, 10))

ax = axes.ravel()

ax[0].imshow(img, cmap=plt.cm.gray)
ax[0].set_title('Imagem Original')

ax[1].imshow(horizontal, cmap=plt.cm.gray)
ax[1].set_title('Imagem horizontal')

ax[2].imshow(vertical, cmap=plt.cm.gray)
ax[2].set_title('Imagem Vertical')

ax[3].imshow(op_sobel, cmap=plt.cm.gray)
ax[3].set_title('Operador de Sobel')

for a in ax:
    a.axis('off')

plt.tight_layout()
plt.show()
```

# PreWitt

```python
horizontal = prewitt_h(img)
vertical = prewitt_v(img)

fig, axes = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True, figsize=(10, 10))

ax = axes.ravel()

ax[0].imshow(img, cmap=plt.cm.gray)
ax[0].set_title('Imagem Original')

ax[1].imshow(horizontal, cmap=plt.cm.gray)
ax[1].set_title('Imagem Horizontal')

ax[2].imshow(vertical, cmap=plt.cm.gray)
ax[2].set_title('Imagem Vertical')

ax[3].imshow(op_sobel, cmap=plt.cm.gray)
ax[3].set_title('Operador de Prewitt')

for a in ax:
    a.axis('off')

plt.tight_layout()
plt.show()
```
