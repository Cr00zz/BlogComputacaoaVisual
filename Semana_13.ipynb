## **toon/cel shading**

Cel shading ou toon shading é um estilo de arte de “renderização” não fotorrealista — que busca a aparência do mundo físico — projetado para fazer gráficos de computador 3D ou obras de arte parecerem tridimensionais, mas criando cores planas em cima de uma cor base, fazendo com que o objeto pareça tridimensional enquanto mantém o efeito 2D no resultado.

A forma mais simples de se obter esse efeito é empregando contornos escuros nos objetos 3D, texturas simples e uma iluminação mais "chapada", isto é, menos realista e mais próxima da iluminação empregada em desenhos animados.

**Quais são as semelhanças e diferenças entre o toon/cel shading e o Gouraud e Phong shading?**

A principal característica semelhante entre esses tipos de shaders é o uso da interpolação entre as cores de uma superfície para a criação de sombras e reflexos de um objeto 3D. E cada um possuí suas particularidades:

O Gouraud shading realiza a interpolação de cores nas áreas triangulares dos objetos tridimensionais, com essa técnica, o objeto fica com suas curvas mais suaves, criando uma conexão mais suave entre elas.

O Phong shading também realiza a interpolação de cores, é uma evolução do Gouraud shading em que as superficies ficam mais suaves, sua principal diferença é que a interpolação de cores ocorre nos vetores normais de uma superfície. Isso permite que ele crie reflexos e sombras mais precisos, além de ser capaz de lidar com superfícies com detalhes e curvas acentuadas.
