# APIs gráficas

As APIs gráficas são uma variante das APIs tradicionais feita especificamente para área de imagem e processamento gráfico 3D, fazendo todo o trabalho de comunicação entre jogos, drivers e a placa de vídeo em si.

As APIs gráficas ficam em uma camada entre as engines (os motores gráficos) dos jogos e o driver. O driver é o responsável pela tradução final, levando ao hardware instruções precisas do que precisa ser feito e entregue de volta aos aplicativos. 

#Vulkan

O Vulkan é uma das mais recentes APIs gráficas. Inicialmente desenvolvido pela AMD, o padrão aberto foi cedido ao grupo Khronos, que continuou o seu desenvolvimento. Ele permite a elaboração de códigos low level, o que garante uma grande vantagem em termos de desempenho.

Seu grande diferencial está na flexibilidade: o desenvolvedor pode escolher o quanto ele pretende se aproximar das camadas mais baixas, tendo acesso tanto a ferramentas mais práticas quanto a funções complexas de alto desempenho.

Outra grande vantagem é a redução do número de Draw Calls, comandos disparados pela CPU para que a GPU renderize objetos na tela, algo que pode se tornar um gargalo conforme mais e mais deles são adicionados à cena.

# Microsoft DirectX

Microsoft DirectX é uma coleção de APIs que tratam de tarefas relacionadas a multimídia, especialmente programação de jogos, para o sistema operacional Microsoft Windows, ou seja, é quem padroniza a comunicação entre software e hardware. Com a padronização de comunicação, o DirectX fornece instruções para que aplicações (jogos, programas gráficos e entre outros, que são escritos para fins de sua utilização), e o respectivo hardware, façam uso dos seus recursos.

Observe o seguinte ao fazer referência ao diagrama de estado do pipeline:

-As tabelas e heaps do descritor podem ser organizados arbitrariamente: SRVs, CBVs e UAVs podem ser referenciados e alocados em qualquer ordem.

-Algumas operações do pipeline são configuráveis. Por exemplo, a fusão de saída normalmente opera em uma base de leitura-modificação-gravação com o destino de renderização e exibições de estêncil de profundidade. No entanto, o pipeline pode ser configurado para que qualquer uma dessas exibições seja somente leitura ou somente gravação.

-Exemplores estáticos não fazem parte dos argumentos raiz, pois são estáticos.
